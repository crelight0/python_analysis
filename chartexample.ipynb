```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# 
plt.style.use('dark_background')

# 
np.random.seed(42)
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
sizes = np.random.uniform(10, 100, 1000)  # 
colors = np.random.uniform(0, 1, 1000)    # 

# 
fig, ax = plt.subplots(figsize=(10, 8))

# 
scatter = ax.scatter(x, y, s=sizes, c=colors, cmap='plasma', alpha=0.6, edgecolors='none')

# 
x_curve = np.linspace(-2, 2, 100)
y_curve = np.sin(x_curve) * 0.5
ax.plot(x_curve, y_curve, color='white', lw=2, alpha=0.8, linestyle='--')

# 
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_title('Cosmic Harmony', fontsize=20, color='white', pad=20)
ax.set_xlabel('X Axis', fontsize=12, color='white')
ax.set_ylabel('Y Axis', fontsize=12, color='white')

# 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')

# 
plt.colorbar(scatter, label='Brightness')

# 
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2)) * np.exp(-0.1 * (X**2 + Y**2))

# 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)

# 
ax.set_title('Rainbow Waves', fontsize=20, pad=20)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

# 
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5, label='Height')

# 
ax.view_init(elev=30, azim=45)

# 
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
theta = np.linspace(0, 2 * np.pi, 1000)
r = np.abs(np.cos(5 * theta))  # 5

# 
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

# 
ax.plot(theta, r, color='crimson', lw=3)
ax.fill(theta, r, color='crimson', alpha=0.3)

# 
ax.set_title('Symmetrical Blossom', fontsize=20, pad=20)
ax.set_rmax(1.2)  # 
ax.grid(True, linestyle='--', alpha=0.5)

# 
ax.set_facecolor('#f0f0f0')  # 
fig.patch.set_facecolor('#f0f0f0')

# 
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# 
np.random.seed(42)
x = np.arange(10)  # 10 
y = np.ones(10)    # Y  1
z = np.random.randint(1, 10, size=10)  # 

# 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

#  3D 
colors = cm.viridis(np.linspace(0, 1, len(x)))  #  viridis 
ax.bar3d(x, y, np.zeros_like(z), 1, 1, z, color=colors, alpha=0.8, shade=True)

# 
ax.set_title('Dreamlike Bars', fontsize=20, color='white', pad=20)
ax.set_xlabel('X Axis', fontsize=12, color='white')
ax.set_ylabel('Y Axis', fontsize=12, color='white')
ax.set_zlabel('Height', fontsize=12, color='white')

# 
fig.patch.set_facecolor('#1a0d2b')  # 
ax.set_facecolor('#2b1a4d')         # 

# 
ax.view_init(elev=25, azim=135)

# 
ax.grid(False)

# 
plt.tight_layout()
plt.show()
```
