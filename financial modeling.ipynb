```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 現金流量預測圖表
months = np.arange(1, 13)
initial_cash = 10000
monthly_income = 10000
monthly_expense = 6000

cash_flow = monthly_income - monthly_expense
ending_cash_balance = [initial_cash + cash_flow * i for i in range(1, 13)]

# 2. 折現現金流（DCF）模型圖表
cash_flows = [100000, 120000, 140000]
discount_rate = 0.10
discounted_cash_flows = [cf / (1 + discount_rate) ** (i + 1) for i, cf in enumerate(cash_flows)]
years = np.arange(1, 4)

# 3. 盈利預測圖表
units_sold = 1000
price_per_unit = 500
cost_per_unit = 300

total_revenue = units_sold * price_per_unit
total_cost = units_sold * cost_per_unit
profit = total_revenue - total_cost

# 圖表設置
fig, ax = plt.subplots(3, 1, figsize=(8, 15))

# 現金流量預測圖表
ax[0].plot(months, ending_cash_balance, label='期末現金餘額', color='b', marker='o')
ax[0].set_title('現金流量預測')
ax[0].set_xlabel('月份')
ax[0].set_ylabel('現金餘額 (元)')
ax[0].grid(True)
ax[0].legend()

# 折現現金流（DCF）模型圖表
ax[1].bar(years, discounted_cash_flows, color='g', alpha=0.6, label='折現後現金流')
ax[1].set_title('折現現金流（DCF）模型')
ax[1].set_xlabel('年份')
ax[1].set_ylabel('折現後現金流 (元)')
ax[1].grid(True)
ax[1].legend()

# 盈利預測圖表
ax[2].bar(['收入', '成本', '盈利'], [total_revenue, total_cost, profit], color=['blue', 'red', 'green'])
ax[2].set_title('盈利預測')
ax[2].set_ylabel('金額 (元)')
ax[2].grid(True)

plt.tight_layout()
plt.show()
```
