```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 
months = np.arange(1, 13)
initial_cash = 10000
monthly_income = 10000
monthly_expense = 6000

cash_flow = monthly_income - monthly_expense
ending_cash_balance = [initial_cash + cash_flow * i for i in range(1, 13)]

# 2. DCF
cash_flows = [100000, 120000, 140000]
discount_rate = 0.10
discounted_cash_flows = [cf / (1 + discount_rate) ** (i + 1) for i, cf in enumerate(cash_flows)]
years = np.arange(1, 4)

# 3. 
units_sold = 1000
price_per_unit = 500
cost_per_unit = 300

total_revenue = units_sold * price_per_unit
total_cost = units_sold * cost_per_unit
profit = total_revenue - total_cost

# 
fig, ax = plt.subplots(3, 1, figsize=(8, 15))

# 
ax[0].plot(months, ending_cash_balance, label='', color='b', marker='o')
ax[0].set_title('')
ax[0].set_xlabel('')
ax[0].set_ylabel(' ()')
ax[0].grid(True)
ax[0].legend()

# DCF
ax[1].bar(years, discounted_cash_flows, color='g', alpha=0.6, label='')
ax[1].set_title('DCF')
ax[1].set_xlabel('')
ax[1].set_ylabel(' ()')
ax[1].grid(True)
ax[1].legend()

# 
ax[2].bar(['', '', ''], [total_revenue, total_cost, profit], color=['blue', 'red', 'green'])
ax[2].set_title('')
ax[2].set_ylabel(' ()')
ax[2].grid(True)

plt.tight_layout()
plt.show()
```
