```python
import requests
import json

url = "https://nghr.gov.taipei/OpenData.aspx?SN=D8ADC700865287FB"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    # 方法 1：用 utf-8-sig 解碼文字
    text = response.text.encode().decode("utf-8-sig")
    data = json.loads(text)
    
    # 方法 2：直接處理位元組（更可靠）
    # data = json.loads(response.content.decode("utf-8-sig"))
    
    # 儲存結果
    with open("output.json", "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    print("資料已下載")
else:
    print(f"請求失敗，狀態碼：{response.status_code}")
```

```python
import json
import pandas as pd

# 假設你的 JSON 已儲存在 output.json
with open("output.json", "r", encoding="utf-8-sig") as f:
    data = json.load(f)

# 整理資料為表格格式
rows = []
for item in data:
    # 提取相關檔案的第一個（如果有多個，可以改進程式碼處理）
    file_title = item["相關檔案"][0]["title"] if item["相關檔案"] else ""
    file_url = item["相關檔案"][0]["url"] if item["相關檔案"] else ""
    
    # 建立每行資料
    row = {
        "DataSN": item["DataSN"],
        "Title": item["title"],
        "Source": item["Source"],
        "FileName": item["FileName"],
        "File_Title": file_title,
        "File_URL": file_url,
        "發布單位": item["發布單位"]
    }
    rows.append(row)

# 轉為 DataFrame
df = pd.DataFrame(rows)

# 調整欄位順序（可根據需求調整）
df = df[["DataSN", "Title", "Source", "FileName", "File_Title", "File_URL", "發布單位"]]

# 顯示表格
print(df)

# 儲存為 CSV
df.to_csv("output_table.csv", index=False, encoding="utf-8-sig")
print("已儲存為 output_table.csv")

# 儲存為 Excel（可選）
df.to_excel("output_table.xlsx", index=False)
print("已儲存為 output_table.xlsx")
```
